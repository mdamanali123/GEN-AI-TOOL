```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
with open('owlcreek.txt','r') as file:
  text = file.read()
```

```python
doc = nlp(text)
```

```python
print(f"Number of tokens in the file: {len(doc)}")
```

```python
sentences = list(doc.sents)

print(f"Number of sentences: {len(sentences)}")
```

```python
print(f"Second Sentence: {sentences[1].text}")
```

```python
for token in doc:
  print(token.text, token.pos_, token.dep_, token.lemma_)

# Get number of tokens
x = len(doc)
print(f"Number of tokens: {x}")
```
